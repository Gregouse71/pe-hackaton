```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
exoplanetes = pd.read_csv('exoplanetes.csv', comment='#')
```

```python
sns.set_theme(style="whitegrid")
```

```python
exoplanetes.plot(x='sy_pnum', y='st_mass', style='.', xlabel='nombre planètes', ylabel='masse étoile', legend=False)
exoplanetes.plot(x='sy_pnum', y='st_dens', style='.', xlabel='nombre planètes', ylabel='densité étoile', legend=False)
exoplanetes.plot(x='sy_pnum', y='st_rad', style='.', xlabel='nombre planètes', ylabel='rayon étoile', legend=False);
```

```python
exoplanetes.plot(x='sy_snum', y='st_mass', style='.', xlabel='nombre étoiles', ylabel='masse étoile', legend=False)
exoplanetes.plot(x='sy_snum', y='st_dens', style='.', xlabel='nombre étoiles', ylabel='densité étoile', legend=False)
exoplanetes.plot(x='sy_snum', y='st_rad', style='.', xlabel='nombre étoiles', ylabel='rayon étoile', legend=False);
```
