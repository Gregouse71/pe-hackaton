```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
exoplanetes = pd.read_csv('exoplanetes.csv', comment='#')
```

```python
sns.set_theme(style="whitegrid")
```

```python
exoplanetes.plot(x='sy_pnum', y='st_mass', style='.', xlabel='nombre plantes', ylabel='masse toile', legend=False)
exoplanetes.plot(x='sy_pnum', y='st_dens', style='.', xlabel='nombre plantes', ylabel='densit toile', legend=False)
exoplanetes.plot(x='sy_pnum', y='st_rad', style='.', xlabel='nombre plantes', ylabel='rayon toile', legend=False);
```

```python
exoplanetes.plot(x='sy_snum', y='st_mass', style='.', xlabel='nombre toiles', ylabel='masse toile', legend=False)
exoplanetes.plot(x='sy_snum', y='st_dens', style='.', xlabel='nombre toiles', ylabel='densit toile', legend=False)
exoplanetes.plot(x='sy_snum', y='st_rad', style='.', xlabel='nombre toiles', ylabel='rayon toile', legend=False);
```
